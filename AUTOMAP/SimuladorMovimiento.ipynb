```python
import pandas as pd
from pprint import pprint

# ======================================================
# Leer las hojas del Excel
# ======================================================

EXCEL_FILE = "chip_layout_normalized.xlsx"

devices_df = pd.read_excel(
    EXCEL_FILE,
    sheet_name="Devices"
)

ports_df = pd.read_excel(
    EXCEL_FILE,
    sheet_name="Ports"
)

measurements_df = pd.read_excel(
    EXCEL_FILE,
    sheet_name="Measurements"
)

# ======================================================
# Construir el objeto CHIP
# ======================================================

chip = {}

# ------------------------------------------------------
# Añadir los dispositivos
# ------------------------------------------------------

for _, row in devices_df.iterrows():

    chip[row["Device"]] = {

        "type": row["Type"],

        "measure": row["Measure"],

        "ports": {}

    }

# ------------------------------------------------------
# Añadir los puertos a cada dispositivo
# ------------------------------------------------------

for _, row in ports_df.iterrows():

    device = row["Device"]

    chip[device]["ports"][row["Port"]] = {

        "direction": row["Direction"],

        "measure": row["Measure"],

        "x": row["X (um)"],

        "y": row["Y (um)"]

    }

# ======================================================
# Mostrar la estructura generada
# ======================================================

print("=" * 70)
print("ESTRUCTURA DEL CHIP")
print("=" * 70)

pprint(chip)

# ======================================================
# Simulación de la rutina de medida
# ======================================================

print("\n")
print("=" * 70)
print("SIMULACIÓN")
print("=" * 70)

for _, measurement in measurements_df.iterrows():

    # --------------------------------------------------
    # ¿La medición está habilitada?
    # --------------------------------------------------

    if measurement["Enabled"] != "YES":
        continue

    device_name = measurement["Device"]
    input_name = measurement["Input"]
    output_name = measurement["Output"]

    # --------------------------------------------------
    # Obtener el dispositivo
    # --------------------------------------------------

    if device_name not in chip:

        print(f"ERROR: {device_name} no existe.")
        continue

    device = chip[device_name]

    # --------------------------------------------------
    # ¿El dispositivo se mide?
    # --------------------------------------------------

    if device["measure"] != "YES":

        print(f"{device_name} ignorado (Measure = NO)")
        continue

    # --------------------------------------------------
    # Obtener puertos
    # --------------------------------------------------

    if input_name not in device["ports"]:

        print(f"ERROR: {input_name} no existe.")
        continue

    if output_name not in device["ports"]:

        print(f"ERROR: {output_name} no existe.")
        continue

    input_port = device["ports"][input_name]
    output_port = device["ports"][output_name]

    # --------------------------------------------------
    # ¿Los puertos están habilitados?
    # --------------------------------------------------

    if input_port["measure"] != "YES":

        print(f"{device_name} - {input_name} ignorado.")
        continue

    if output_port["measure"] != "YES":

        print(f"{device_name} - {output_name} ignorado.")
        continue

    # --------------------------------------------------
    # Simulación
    # --------------------------------------------------

    print("=" * 60)

    print(f"Dispositivo : {device_name}")
    print(f"Tipo        : {device['type']}")

    print()

    print(
        f"Entrada ({input_name}) "
        f"-> ({input_port['x']:.1f}, {input_port['y']:.1f})"
    )

    print(
        f"Salida  ({output_name}) "
        f"-> ({output_port['x']:.1f}, {output_port['y']:.1f})"
    )

    print()

    print(
        f"Mover fibra izquierda  -> "
        f"({input_port['x']:.1f}, {input_port['y']:.1f})"
    )

    print(
        f"Mover fibra derecha    -> "
        f"({output_port['x']:.1f}, {output_port['y']:.1f})"
    )

    print()

    print("Alinear entrada...")
    print("Alinear salida...")
    print("Medir potencia...")

print("\nProceso finalizado.")
```

```python
import pandas as pd

# ======================================================
# Leer las hojas del Excel
# ======================================================

devices_df = pd.read_excel(
    "chip_layout_normalized.xlsx",
    sheet_name="Devices"
)

ports_df = pd.read_excel(
    "chip_layout_normalized.xlsx",
    sheet_name="Ports"
)

measurements_df = pd.read_excel(
    "chip_layout_normalized.xlsx",
    sheet_name="Measurements"
)

# ======================================================
# Simulación del recorrido
# ======================================================

for _, measurement in measurements_df.iterrows():

    # Saltar las mediciones deshabilitadas
    if measurement["Enabled"] != "YES":
        continue

    device = measurement["Device"]
    input_port = measurement["Input"]
    output_port = measurement["Output"]

    # Buscar coordenadas del puerto de entrada
    input_data = ports_df[
        (ports_df["Device"] == device) &
        (ports_df["Port"] == input_port)
    ].iloc[0]

    # Buscar coordenadas del puerto de salida
    output_data = ports_df[
        (ports_df["Device"] == device) &
        (ports_df["Port"] == output_port)
    ].iloc[0]

    print("=" * 50)
    print(f"Dispositivo : {device}")

    print(
        f"Entrada : {input_port}"
        f" -> ({input_data['X (um)']}, {input_data['Y (um)']})"
    )

    print(
        f"Salida  : {output_port}"
        f" -> ({output_data['X (um)']}, {output_data['Y (um)']})"
    )

    print()

    print(f"Mover fibra izquierda  a ({input_data['X (um)']}, {input_data['Y (um)']})")
    print(f"Mover fibra derecha    a ({output_data['X (um)']}, {output_data['Y (um)']})")

    print("Alinear...")
    print("Medir potencia...")
    print()
```
