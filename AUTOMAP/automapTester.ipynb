```python
import pandas as pd
from pprint import pprint
from openpyxl import load_workbook
```

```python
# ======================================================
# Leer las hojas del Excel
# ======================================================

devices_df = pd.read_excel(
    "chip_layout_normalized.xlsx",
    sheet_name="Devices"
)

ports_df = pd.read_excel(
    "chip_layout_normalized.xlsx",
    sheet_name="Ports"
)

print(devices_df)

print()

print(ports_df)
```

```python
# ======================================================
# Construir la lista de dispositivos
# ======================================================

devices = []

for _, device_row in devices_df.iterrows():

    # Buscar los puertos que pertenecen a este dispositivo
    device_ports = ports_df[
        ports_df["Device"] == device_row["Device"]
    ]

    # Separar entradas y salidas
    inputs = device_ports[
        device_ports["Direction"] == "IN"
    ]

    outputs = device_ports[
        device_ports["Direction"] == "OUT"
    ]

    # Crear el diccionario del dispositivo
    device = {

        "name": device_row["Device"],

        "type": device_row["Type"],

        "inputs": [],

        "outputs": []

    }

    # Añadir las entradas
    for _, port in inputs.iterrows():

        device["inputs"].append({

            "port": port["Port"],

            "x": port["X (um)"],

            "y": port["Y (um)"]

        })

    # Añadir las salidas
    for _, port in outputs.iterrows():

        device["outputs"].append({

            "port": port["Port"],

            "x": port["X (um)"],

            "y": port["Y (um)"]

        })

    # Guardar el dispositivo
    devices.append(device)
```

```python
pprint(devices)
```

```python
# ======================================================
# Mostrar la información de forma amigable
# ======================================================

for device in devices:

    print("\n" + "=" * 40)

    print(f"Dispositivo : {device['name']}")
    print(f"Tipo        : {device['type']}")

    print("\nEntradas:")

    for port in device["inputs"]:

        print(
            f"   {port['port']:>4}  ->  "
            f"({port['x']:.1f}, {port['y']:.1f})"
        )

    print("\nSalidas:")

    for port in device["outputs"]:

        print(
            f"   {port['port']:>4}  ->  "
            f"({port['x']:.1f}, {port['y']:.1f})"
        )
```

Necesitamos añadir una nueva columna que le indique a la maquina la relación de puertos que tendra en cuenta al momento de tomar las medidas 

```python
excel_file = "chip_layout_normalized.xlsx"

wb = load_workbook(excel_file)
```

```python
# ======================================================
# Crear la hoja Measurements
# ======================================================

if "Measurements" in wb.sheetnames:
    del wb["Measurements"]

ws = wb.create_sheet("Measurements")

# Encabezados
ws.append([
    "Device",
    "Input",
    "Output",
    "Enabled",
    "Status"
])
```

```python
# ======================================================
# Generar automáticamente las mediciones
# ======================================================

for device in devices:

    for input_port in device["inputs"]:

        for output_port in device["outputs"]:

            ws.append([

                device["name"],

                input_port["port"],

                output_port["port"],

                "YES",

                "Pending"

            ])

# ======================================================
# Guardar
# ======================================================

wb.save(excel_file)

print("Hoja Measurements creada correctamente.")
```

Tiene limitaciones con respecto a la forma en la que se manejan los puertos, el usuario tendria que enumerar todos los puertos a la entrada y a la salida lo que resulta poco practico, es mejor permitir que cada dispositivo tenga su propio manejo de los puertos 

```python
import pandas as pd

# Leer únicamente la hoja "Devices"
df = pd.read_excel("chip_layout_example.xlsx", sheet_name="Devices")

print(df)
```

```python
devices = []

for _, row in df.iterrows():

    outputs = str(row["OutputPorts"]).split(";")

    device = {
        "name": row["Device"],
        "type": row["Type"],
        "input": row["InputPort"],
        "outputs": outputs,
    }

    devices.append(device)
```

```python
devices 
```

```python
for device in devices:

    print("=" * 40)

    print(f"Dispositivo : {device['name']}")
    print(f"Tipo        : {device['type']}")
    print(f"Entrada     : {device['input']}")

    print("Salidas:")

    for i, output in enumerate(device["outputs"], start=1):

        print(f"   Puerto {i:2d} -> {output}")
```
